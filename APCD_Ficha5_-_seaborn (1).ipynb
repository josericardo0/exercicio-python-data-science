# Gráficos relacionais (numerical x numerical)

## tips | relplot | total_bill vs tip - day

```python
# Gerando um gráfico relacional (total_bill vs tip) diferenciado por dia da semana ('day')
sns.relplot(x='total_bill', y='tip', hue='day', data=tips, palette='viridis', height=6, aspect=1.2)

# Adicionando rótulos para os eixos e título para o gráfico.
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.title('Total Bill vs Tip by Day')

# Exibindo o gráfico
plt.show()
```

```python

# Gerando um gráfico de dispersão com diferentes cores
sns.scatterplot(x='date', y='price', data=house, palette='coolwarm', alpha=0.7)

# Definindo os rótulos para os eixos e o título para o gráfico 
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatterplot: Age vs Charges')

# Exibindo o gráfico
plt.show()
```

# Gráficos categóricos (categorical x numerical)

## stroke | stripplot | stroke vs age

```python
# Criando um gráfico categórico com stripplot para 'stroke' versus 'age'
sns.stripplot(x='stroke', y='age', data=stroke, jitter=True, palette='Set1', alpha=0.7)

# Adicionando rótulos e título ao gráfico.
plt.xlabel('Stroke')
plt.ylabel('Age')
plt.title('Stripplot: Stroke vs Age')

# Exibindo o gráfico
plt.show()
```

## stroke | swarmplot | stroke vs age

```python

# Trazendo apenas 100 registros devido ao tamanho do dataset
data_exemplo = stroke.sample(n=100, random_state=42)

# Criando um gráfico categórico com swarmplot para 'stroke' versus 'age'
sns.swarmplot(x='stroke', y='age', data=data_exemplo, palette='Set1', alpha=0.7)

# Adicionando rótulos e título ao gráfico.
plt.xlabel('Stroke')
plt.ylabel('Age')
plt.title('Swarmplot: Stroke vs Age')

# Exibindo o gráfico
plt.show()
```

## stroke | boxplot | stroke vs age

```python
# Criando um gráfico boxplot para 'stroke' versus 'age'
sns.boxplot(x='stroke', y='age', data=stroke, palette='Set1')

# Adicionando rótulos e título ao gráfico.
plt.xlabel('Stroke')
plt.ylabel('Age')
plt.title('Boxplot: Stroke vs Age')

# Exibindo o gráfico.
plt.show()
```

## stroke | boxenplot | stroke vs age

```python
# Criando um gráfico boxenplot para 'stroke' versus 'age'
sns.boxenplot(x='stroke', y='age', data=stroke, palette='Set1')

# Adicionando rótulos e título
plt.xlabel('Stroke')
plt.ylabel('Age')
plt.title('Boxenplot: Stroke vs Age')

# Exibindo o gráfico
plt.show()
```

## stroke | violinplot | stroke vs age

```python
# Criando um gráfico violinplot para 'stroke' versus 'age'
sns.violinplot(x='stroke', y='age', data=stroke, palette='Set1')

# Adicionando rótulos e título
plt.xlabel('Stroke')
plt.ylabel('Age')
plt.title('Violinplot: Stroke vs Age')

# Exibindo o gráfico
plt.show()
```

# Gráficos de regressão (numerical x numerical)

## stroke | lmplot | age vs bmi  - stroke

```python
# Criando um gráfico lmplot para 'age' versus 'bmi' através de 'stroke'
sns.lmplot(x='age', y='bmi', hue='stroke', data=stroke, palette='Set1', scatter_kws={'alpha':0.7})

# Adicionando rótulos e título
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('LMplot: Age vs BMI (colored by Stroke)')

# Exibindo o gráfico
plt.show()
```

## tips | lmplot | total_bill vs tip - day

```python

# Criando um gráfico lmplot para 'total_bill' através de 'tip'
sns.lmplot(x='total_bill', y='tip', data=tips, hue='day', aspect=1.5, height=6)

# Adicionando título ao gráfico.
plt.title('Relação entre Total Bill e Tip por Dia')

# Exibindo o gráfico.
plt.show()
```

## tips | lmplot | total_bill vs tip

```python
# Criando um gráfico lmplot para 'total_bill' versus 'tip' através de'day'
sns.lmplot(x='total_bill', y='tip', hue='day', data=tips, palette='Set1', scatter_kws={'alpha':0.7})

# Adicionando rótulos e título ao gráfico.
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('LMplot: Total Bill vs Tip (colored by Day)')

# Exibindo o gráfico
plt.show()
```

## tips | residplot | total_bill vs tip

```python
# Criando um gráfico residplot para 'total_bill' versus 'tip'
sns.residplot(x='total_bill', y='tip', data=tips, scatter_kws={'alpha':0.7})

# Adicionando rótulos e título ao gráfico.
plt.xlabel('Total Bill')
plt.ylabel('Residuals')
plt.title('Residplot: Total Bill vs Tip')

# Exibindo o gráfico
plt.show()
```

## insurance | lmplot | age vs charges   -  smoker

```python
# Criando um gráfico lmplot para 'age' versus 'charges' diferenciado por 'smoker'
sns.lmplot(x='age', y='charges', hue='smoker', data=insurance, palette='Set1', scatter_kws={'alpha':0.7})

# Adicionando rótulos e título ao gráfico.
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('LMplot: Age vs Charges (colored by Smoker)')

# Exibindo o gráfico
plt.show()
```

## insurance | regplot | age vs charges

```python
# Criando um gráfico regplot para 'age' versus 'charges'
sns.regplot(x='age', y='charges', data=insurance, scatter_kws={'alpha':0.7})

# Adicionando rótulos e título ao gráfico.
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Regplot: Age vs Charges')

# Exibindo o gráfico.
plt.show()
```
